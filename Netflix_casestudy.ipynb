```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\netflix_dataset.csv')

# Count the occurrences of each type
type_counts = df['type'].value_counts()

# Calculate percentages
total_count = type_counts.sum()
percentages = (type_counts / total_count) * 100

# Print the percentages rounded to one decimal point
tv_shows_percentage = percentages.get('TV Show', 0)
movies_percentage = percentages.get('Movie', 0)
print(f'({movies_percentage:.1f}, {tv_shows_percentage:.1f})')

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Percentage of TV Shows and Movies')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\netflix_dataset.csv')

# Count the number of movies and TV shows by country
country_counts = df['country'].value_counts()

# Get the top 10 countries with the most content
top_10_countries = country_counts.head(10)

# Plot the bar chart
plt.figure(figsize=(12, 8))
top_10_countries.plot(kind='bar', color='skyblue')
plt.xlabel('Country')
plt.ylabel('Number of Movies and TV Shows')
plt.title('Top 10 Countries with the Most Content')
plt.xticks(rotation=45)
plt.show()

# Get the top country and its total content
top_country = top_10_countries.idxmax()
total_content = top_10_countries.max()

# Print the output in the required format
print(f'({top_country}, {total_content})')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\netflix_dataset.csv')

# Filter the dataset for content in India
india_df = df[df['country'] == 'India']

# Analyze the content type
content_type_counts = india_df['type'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(content_type_counts, labels=content_type_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Distribution of Content Type in India')
plt.show()

# Determine the most popular content type
most_popular_content_type = content_type_counts.idxmax()

# Analyze the rating type
rating_counts = india_df['rating'].value_counts()
plt.figure(figsize=(12, 8))
rating_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Rating Type')
plt.ylabel('Number of Titles')
plt.title('Distribution of Rating Types in India')
plt.xticks(rotation=45)
plt.show()

# Determine the most popular rating type
most_popular_rating_type = rating_counts.idxmax()

# Print the output in the required format
print(f'({most_popular_content_type}, {most_popular_rating_type})')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\netflix_dataset.csv')

# Assuming there is a 'Date_added' column in the dataset
# Convert the 'Date_added' column to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract the year from 'Date_added' and create a 'Year' column
df['Year'] = df['date_added'].dt.year

# Count the number of content added per year
content_per_year = df['Year'].value_counts().sort_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
content_per_year.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of Content Added')
plt.title('Content Added Per Year on Netflix')
plt.show()

# Determine the year with the most content added
year_most_content = content_per_year.idxmax()
print(year_most_content)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('C:\\Users\\Prajwal\\Desktop\\Datasets\\netflix_dataset.csv')

# Filter the dataset for movies only
movies_df = df[df['type'] == 'Movie']

# Convert 'duration' to numeric after removing ' min' from the column
movies_df['duration'] = movies_df['duration'].str.replace(' min', '').astype(float)

# Find the movie with the longest duration
longest_movie = movies_df.loc[movies_df['duration'].idxmax()]

# Extract the title and release year of the longest movie
longest_movie_title = longest_movie['title']
release_year = longest_movie['release_year']

# Print the result in the required format
print(f"('{longest_movie_title}', {release_year})")
```
